# data cleaning

```python
df.isnull().sum()
```

```python
df["Age"].fillna(df["Age"].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
missing_cabin = df["Cabin"].isnull().mean()
missing_cabin
```

```python
df["has_cabin"] = np.where(df["Cabin"].notnull(), "yes","no")
```

```python
df
```

```python
threshold = 0.7
if missing_cabin > threshold:
  df["Cabin"] = df["Cabin"].fillna("unknown")
```

```python
df = df.drop("Cabin", axis=1)
```
